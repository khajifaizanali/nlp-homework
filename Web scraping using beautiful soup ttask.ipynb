## importing required modules

```python
import bs4
import requests
```

```python
URL = "http://www.values.com/inspirational-quotes"
```

```python
from bs4 import BeautifulSoup
```

```python
r = requests.get(URL) 
```

```python
l=bs4.BeautifulSoup(r.content,'html.parser'); # parsing the html content
```

```python
print(l.title.string)
```

```python
# using prettify to get the html data
print(l.prettify())
```

```python
#getting required sections
table = l.find('section', attrs = {'class':'half-section bg-extra-dark-gray half-section'}) 
```
